```python
import matplotlib.pyplot as plt
import numpy as np

# First 
x= np.array([1, 9])
y= np.array([3, 10])
plt.subplot(2, 3, 1)
plt.plot(x, y, 'o')
plt.xlabel("X Axis",color='blue')
plt.ylabel("Y Axis",color='blue')
plt.title("First Plot")

# Second 
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([3, 2, 3, 2, 3])
y2 = np.array([2, 3, 2, 3, 2])

plt.subplot(2,3,2)
plt.plot(x, y1, color='orange', linestyle='dashed', linewidth=2.5)
plt.plot(x, y2, color='blue', marker='*')
plt.xlim(0.5, 5.5)
plt.ylim(1.5, 3.5)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Third
x = np.array([1, 2, 3, 7])
plt.subplot(3, 3, 3)  
plt.plot(x, marker='o', mec='black', mfc='yellow')


# Fourth
x = np.array([1, 6])
plt.subplot(2, 3, 4)  
plt.plot(x)
plt.grid()


# Fifth
plt.subplot(2,3,5)
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
colors = np.random.rand(len(x))  # Define colors

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.title("Fifth Plot")

# Sixth
plt.subplot(2,3,6)
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, color="red", width=0.1)
plt.suptitle("GRAPH")
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 20, 20, 25])
myexplode = [0.2, 0, 0, 0]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y,startangle = 90,explode = myexplode,shadow = True,colors = mycolors,labels = mylabels)
plt.legend()
plt.show()
```
